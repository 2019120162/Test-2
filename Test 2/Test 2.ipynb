```python
import pandas as pd
import numpy as np
```

```python
bint_df = pd.read_csv("BINT.csv")
```

```python
bint_df
```

```python
bint_df_drop = bint_df.drop(["ETHNICITYID", "CITYID", "DISTRICTID", "DEGREEID", "DEGREE", "FEEDERID", "COMMENTS", "CODE", "PROGRAM", "BREAK"], axis=1)
```

```python
bint_df_clean =bint_df_drop.drop([8443],axis=0)
```

```python
bint_df_clean.nunique()
```

```python
bint_df_clean_unique = bint_df_clean.drop_duplicates(subset="STUID")
```

```python
bint_df_clean_unique
```

```python
#Number of students

bint_df_clean_unique["STUID"].value_counts()
```

```python
student_count = bint_df_clean_unique["STUID"].value_counts().sum()
```

```python
student_count
```

```python
#Number of students who graduated

students_graduated_df = bint_df_clean_unique[bint_df_clean_unique["PROGRAMSTATUS"] == "Graduated"]
```

```python
students_graduated = students_graduated_df["STUID"].nunique()
```

```python
students_graduated
```

```python
#Gender breakdown

bint_df_clean_unique.groupby("GENDER")["GENDER"].count()
```

```python
#Ethnicity breakdown

bint_df_clean_unique.groupby("ETHNICITY")["ETHNICITY"].count().sort_values(ascending = False)
```

```python
#Grade breakdown

bint_df_clean.groupby("COURSEGRADE")["COURSEGRADE"].count()
```

```python
#Average number of years to graduate from BINT

#This creates a copy of the DataFrame without affecting the original DataFrame.
bint_df_clean_unique_copy = bint_df_clean_unique.copy()

#Convert PROGRAMSTART and PROGRAMEND columns to datetime
bint_df_clean_unique_copy["PROGRAMSTART"] = pd.to_datetime(bint_df_clean_unique_copy["PROGRAMSTART"])
bint_df_clean_unique_copy["PROGRAMEND"] = pd.to_datetime(bint_df_clean_unique_copy["PROGRAMEND"])

#Calculate difference between PROGRAMEND and PROGRAMSTART columns
bint_df_clean_unique_copy["PROGRAMDURATION"] = (bint_df_clean_unique_copy["PROGRAMEND"] - bint_df_clean_unique_copy["PROGRAMSTART"]).dt.days / 365

# Calculate average number of years
average_duration = bint_df_clean_unique_copy["PROGRAMDURATION"].mean()
```

```python
#Average in years

average_duration
```
